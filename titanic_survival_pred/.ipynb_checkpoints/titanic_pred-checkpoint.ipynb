```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
sns.heatmap(train.isna(),cbar=False,yticklabels=False)
```

```python
train = train.drop('Cabin',axis=1)
test = test.drop('Cabin',axis=1)
```

```python
sns.heatmap(train.isna(),cbar=False,yticklabels=False)
```

```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='Pclass',y='Age',data=train,palette='winter')
```

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):

        if Pclass == 1:
            return 37

        elif Pclass == 2:
            return 29

        else:
            return 24

    else:
        return Age
```

```python
train['Age'] = train[['Age','Pclass']].apply(impute_age,axis=1)
test['Age'] = test[['Age','Pclass']].apply(impute_age,axis=1)
```

```python
sns.heatmap(train.isna(),cbar=False,yticklabels=False)
```

```python
sex = pd.get_dummies(train['Sex'],drop_first=True)
embark = pd.get_dummies(train['Embarked'],drop_first=True)

sex_t = pd.get_dummies(test['Sex'],drop_first=True)
embark_t = pd.get_dummies(test['Embarked'],drop_first=True)
```

```python
train.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace=True)
test.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace=True)
```

```python
train = pd.concat([train,sex,embark],axis=1)
test = pd.concat([test,sex_t,embark_t],axis=1)
```

```python
train.head()
```

```python
test.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
dtree = DecisionTreeClassifier()
```

```python
X_train = train.drop(['Survived','PassengerId'],axis=1)
y_train = train['Survived']

X_test = test.drop('PassengerId',axis=1)
```

```python
dtree.fit(X_train,y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gbn = GaussianNB()
```

```python
y_pred = gbn.fit(X_train,y_train).predict(X_test)
```

```python
print("Number of mislabeled points out of total %d points : %d" %(X_test.shape[0], (y_test != y_pred).sum()))
```
